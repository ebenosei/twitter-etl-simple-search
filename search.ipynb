```python
#Import libraries

import tweepy
from tweepy import OAuthHandler
import json
from datetime import datetime
import os
import sys
import pandas as pd
```

```python
#Copy and paste API keys from twitter dev account

consumer_key = '**********'
consumer_secret = '**********'
access_token = '*****-********'
access_secret = '*********'
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#Mikel Arterta is currently the manager of English football club, Arsenal FC. He is currently in the bottom half of the league and this has caused a call for his dismissal amongst some fans, hence the #ArtetaOut hashtag

text_query = '#artetaout'

tweets = tweepy.Cursor(api.search, q=text_query, lang='en').items(1500)



ArtetaOut = []
for tweet in tweets:
       ArtetaOut.append(
         {"Date": tweet.created_at, "username": tweet.user.name, "screen_name": tweet.user.screen_name, "user_location": tweet.user.location, "user_friends_count": tweet.user.friends_count, "verified?": tweet.user.verified, "user_bio": tweet.user.description, "followers_count": tweet.user.followers_count, "Tweet": tweet.text, "likes" : tweet.favorite_count, "retweets": tweet.retweet_count, "source": tweet.source, "tweet_id": tweet.id, "tweet_coordinates": tweet.coordinates, "hashtags" :tweet.entities['hashtags']})



ArtetaOut = pd.DataFrame(ArtetaOut)
ArtetaOut
```

```python
#convert ArtetaOut data frame to .csv
ArtetaOut.to_csv(f'{datetime.today().strftime("%Y-%m-%d %H-%M-%S")}.csv')
```
